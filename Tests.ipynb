```python
import sys
import datetime
import numpy as np
from tqdm import tqdm
from stable_baselines3 import DQN
from utils import evaluate as ev
from utils import evaluate_extended as ev2
import gym
import sqli_sim
```

```python
env_name = 'sqli_sim-v0'
```

```python
n_simulations = 1
n_episodes_training = 10 ** 6
n_episodes_test = 10 ** 2
```

## Running the simulations on escape = 5 columns = 5

```python
env = gym.make(env_name, escapes=5, columns=5)
model1 = [DQN('MlpPolicy', env, verbose=1) for i in range(n_simulations)]
print(model1)
```

```python
train_data = np.zeros((n_simulations, 3, n_episodes_training))
test_data = np.zeros((n_simulations, 3, n_episodes_test))

for i in tqdm(range(n_simulations)):
    model1[i].learn(total_timesteps=10 ** 6)
    model1[i].save(str(i) + 'ignore_test1')
```

```python
for i in tqdm(range(n_simulations)):
    mean_reward, _ = ev.evaluate_model(model1[i].load(str(i) + 'ignore_test1'), env, num_steps=10 ** 6)
    print('dqn Mean reward: {0}'.format(mean_reward))
```
